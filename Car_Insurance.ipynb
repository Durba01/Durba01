```python
import pandas as pd
data = pd.read_csv("/content/Car_Insurance.csv")
```

```python
data.head()
```

```python
# Getting info from the data 
data.info()
```

```python
# Checking missing values 
missing_values = data.isnull().sum()
print(missing_values)
```

```python
from pandas.core.ops.array_ops import missing
# drop the rows with missing values 
cleaned_data = missing_values.dropna()
cleaned_data
```

```python
# Getting numerical values 
numerical_columns = data.select_dtypes(exclude=['object','category','datetime'])
numerical_columns.sum()
```

```python
# Get the mean, median and quartile  of the numerical values 
numerical_columns.describe()
```

```python
# job categories 
num_categories = data['Job'].nunique()
num_categories
```

```python
# Most frequent job
job_counts = data['Job'].value_counts()
most_frequent = job_counts.idxmax()
most_frequent 
```

```python
data[['CallStart','CallEnd']]
```

```python
from datetime import datetime
#convert the start and end time columns to datetime objects
data['CallStart'] = pd.to_datetime(data['CallStart'], format='%H:%M:%S')
data['CallEnd'] = pd.to_datetime(data['CallEnd'], format='%H:%M:%S')

#create a new column with the duration in seconds
data['Duration'] = (data['CallEnd'] - data['CallStart']).dt.total_seconds()
```

```python
data['Duration']
```

```python
#calculate the average duration
average_duration = data['Duration'].mean()
average_duration
```

```python
#create a condition to filter the calls made during first half of the year
condition = (data['CallStart'].dt.month >= 1) & (data['CallStart'].dt.month <= 6)
df_first_half = data[condition]
#calculate the most common way of communication
most_common_communication = df_first_half['Communication'].mode()
most_common_communication
```
